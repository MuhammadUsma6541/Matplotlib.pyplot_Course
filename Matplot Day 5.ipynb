# Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
area=[1,2,8,4,5]
area1=[3,5,6,7,8]
area2=[9,2,4,6,7]
l=['area','area1','area2']
plt.stackplot(x,area,area1,area2,labels=l,colors=['r','b','m'],baseline='sym')
plt.legend(loc=2)
plt.show()
```

# Step Plot

```python
x=[1,2,3,4,5]
y=[5,6,3,4,5]
plt.step(x,y,color='r',marker='o',ms=10,mfc='b',label="Python")

plt.grid()
plt.title('Python')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

# Fill_Between Plot

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.title('Python')
plt.show()
```
