# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x=['python','c','c++','java']
y=[85,70,72,80]
c=['y','b','m','r']
plt.xlabel('Lnaguages',fontsize=15)
plt.ylabel('popularity',fontsize=15)
plt.title('Popularity Of Languages',fontsize=15)
plt.bar(x,y,width=0.4,color=c,align='edge')
plt.show()
```
