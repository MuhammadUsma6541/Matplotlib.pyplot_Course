# Pie PLot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['c','c++','java','Python']
ex=[0.2,0.0,0.0,0.0]
c=['r','g','b','y']
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,radius=0.75,labeldistance=1.1
        ,startangle=0,textprops={'fontsize':10},counterclock=False,wedgeprops={'linewidth':4}
       )
plt.title('Language Ranking')
plt.legend(loc=2)
plt.show()
```
