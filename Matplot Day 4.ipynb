# Stem Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,2,4,5,6,7]
plt.stem(x,y,linefmt=":",markerfmt='ro',bottom=0,basefmt='g',label="Python",use_line_collection=False,orientation='')
plt.legend()
plt.show()
```

# Box Plot

```python

x=[10,20,30,40,50,60,70]
plt.boxplot(x,notch=True,labels=['Python'])
plt.legend()
plt.show()
```
