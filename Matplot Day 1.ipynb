# Bar Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=["Python",'Java','React']
y=[90,88,78]
z=[22,33,12]
```

```python
plt.xlabel('language',fontsize=15)
plt.ylabel('rating',fontsize=15)
plt.title('Usama',fontsize=15)
c=['y','b','m']
plt.bar(x,y,width=0.2,color='y',align='edge',label='Usama',linewidth=5)
plt.bar(x,z,width=0.2,color='r',align='edge',edgecolor='r',linewidth=6,label='Usama1')
plt.legend()
plt.show()
```

```python
plt.xlabel('language',fontsize=15)
plt.ylabel('rating',fontsize=15)
plt.title('Usama',fontsize=15)

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.bar(p,y,width,color='y',align='edge',label='Usama',linewidth=5)
plt.bar(p1,z,width,color='r',align='edge',linewidth=5,label='Usama1')

plt.xticks(p+width,x)
plt.legend(loc='best')
plt.show()
```


# Scatter Plot

```python
day=[1,2,3,4,5,5]
number=[3,4,56,7,8,90]
plt.title('scatter plot',fontsize=15,)
plt.xlabel('Day',fontsize=20)
plt.ylabel('Number',fontsize=20)
plt.scatter(day,number,color='r')
```

```python
day=[1,2,3,4,5,5]
number=[3,4,56,7,8,90]
color=[10,20,30,40,45,99]
sizes=[100,200,300,250,350,400]
plt.title('scatter plot',fontsize=15,)
plt.xlabel('Day',fontsize=20)
plt.ylabel('Number',fontsize=20)
plt.scatter(day,number,c=color,s=sizes,cmap='viridis')
plt.colorbar()
plt.show()
```
